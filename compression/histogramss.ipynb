```python
import pandas
from matplotlib import pyplot as plt
from matplotlib import rcParams, patheffects
import numpy as np
```

```python
deltas = pandas.read_csv("deltas.csv", index_col=False, header=None, names=["percentage", "value"]).sort_values("value")
distances = pandas.read_csv("distances.csv", index_col=False, header=None, names=["percentage", "value"]).sort_values("value")

distances
```

```python
plt.xkcd()
# for some reason xkcd style adds a massive white border to everything, which we remove here
rcParams.update({'path.effects': []})
fig, axes = plt.subplots(2,1)
ax, distance_ax = axes
fig.subplots_adjust(hspace=0.5)

fig.set_dpi(500)
ax.grid(axis='x', which="both", linewidth=0.2)
ax.set_xticks(np.arange(160), minor=True)
distance_ax.grid(axis='x', which="both", linewidth=0.2)
distance_ax.set_xticks(np.arange(80), minor=True)
ax.bar(x=deltas["value"], height=deltas["percentage"], width=0.7)
distance_ax.bar(x=distances["value"], height=distances["percentage"], width=0.7)
distance_ax.set_title("DISTANCES")
ax.set_title("DELTAS")
```
